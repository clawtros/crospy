```python
inputs = """7
UserA: 41.04
UserB: 7.06
UserC: 20.63
UserD: 54.28
UserE: 12.59
UserF: 31.17
UserG: 63.04"""
lines = inputs.split("\n")

def format_input(line):
    a, b = line.split(": ")
    return a, float(b)

l = sorted([format_input(i) for i in lines[1:int(inputs[0]) + 1]], 
            cmp=lambda x,y: cmp(x[1], y[1]))
out = []
last = 0
for c, t in l:
    out.append((c, int(60 + (last - t))))
    last = t
print "\n".join("%s: %d" % (c, t) for c,t in out)
```

```python
def recsum(s, n, acc = []):
    if s < n/2:
        result = [[s, n-s]]
    else:
        return []
    acc.append(s)
    for i in range(s+1, n/2 +1):
        result.append(recsum(i, n - sum(acc)- i, acc))
    for i in range(s+1, n/2 + 1):
        result.append(recsum(i, n, acc))
    return result

recsum(0, 15)
```
